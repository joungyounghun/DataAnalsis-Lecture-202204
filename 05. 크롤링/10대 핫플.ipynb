```python
map = folium.Map(location=[37.531, 126.986], tiles='OpenStreetMap', zoom_start=12)
folium.Marker(
    location=[37.500344, 127.027790],
    popup=folium.Popup('맛집', max_width=200),  #1
    tooltip='아트 몬스터 강남역점'
).add_to(map)
folium.Marker(
    location=[37.578365, 126.985818],
    popup=folium.Popup('전통주 전문점', max_width=200),
    tooltip='전통주 갤러리',
    icon=folium.Icon(color='red', icon='star')  #2
).add_to(map)
folium.Marker(
    location=[37.522914 ,127.022376],
    popup=folium.Popup('카페', max_width=200),
    tooltip='더 앨리',
    icon=folium.Icon(icon='glyphicon-glass') # 3
).add_to(map)
folium.Marker(
    location=[37.519673 , 127.022703],
    popup=folium.Popup('카페', max_width=200),
    tooltip='카페 키츠네',
    icon=folium.Icon(color='purple', icon='glyphicon-euro')  #4
).add_to(map)
folium.Marker(
    location=[37.522406 ,127.048823],
    popup=folium.Popup('맛집', max_width=200),
    tooltip='티컬렉티브',
    icon=folium.Icon(color='orange', icon='glyphicon-heart') #5
).add_to(map)
folium.Marker(
    location=[37.509994,127.051850],
    popup=folium.Popup('맛집', max_width=200),
    tooltip='스물다섯',
    icon=folium.Icon(color='darkred', icon='glyphicon-user')  #6
).add_to(map)
folium.Marker(
    location=[37.505746,127.056223],
    popup=folium.Popup('카페', max_width=200),
    tooltip='테라로사 포스코센터점',
    icon=folium.Icon(color='lightred', icon='glyphicon-off') #7
).add_to(map)
folium.Marker(
    location=[37.503318,127.065463],
    popup=folium.Popup('놀이시설', max_width=200),
    tooltip='바운스트램폴린파크 삼성센터',
    icon=folium.Icon(color='beige', icon='glyphicon-road') #8
).add_to(map)
folium.Marker(
    location=[37.502244,127.25996],
    popup=folium.Popup('식당 (지금은 폐업)', max_width=200),
    tooltip='핫스타',
    icon=folium.Icon(color='darkblue', icon='glyphicon-remove-circle') #9
).add_to(map)
folium.Marker(
    location=[37.510387,127.057332],
    popup=folium.Popup('돈까스 맛집', max_width=200),
    tooltip='등심 팩토리',
    icon=folium.Icon(color='darkgreen', icon='glyphicon-piggy-bank')#10
).add_to(map)
folium.Circle(
    location=[37.517236, 127.047324], tooltip='강남구청',
    radius=300, color='crimson', fill=True
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">10대 핫플 표기</h3>'    
map.get_root().html.add_child(folium.Element(title_html)) 
map
```
